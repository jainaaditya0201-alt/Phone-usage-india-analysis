```python
import pandas as pd

# Load CSV
df = pd.read_csv("phone_usage_india.csv")

# View first few rows
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()   # check missing values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style="whitegrid", palette="pastel")
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Phone Brand", data=df)
plt.title("Phone Brand Distribution in India")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Phone Brand", y="Screen Time (hrs/day)", data=df)
plt.title("Usage Hours by Phone Brand")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df["Screen Time (hrs/day)"], fill=True)
plt.title("Distribution of Phone Usage Hours")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="Age_Group", y="Usage_Hours", data=df)
plt.title("Phone Usage by Age Group")
plt.show()
```

```python
# Create Age_Group column by binning the 'Age' column
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 18, 35, 60, 100], labels=['<18', '18-35', '36-60', '60+'])
display(df.head())
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="Age_Group", y="Screen Time (hrs/day)", data=df)
plt.title("Phone Usage by Age Group")
plt.show()
```
